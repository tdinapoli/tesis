```python
df = pd.read_pickle("/home/tomi/Documents/facultad/tesis/git/data/oscuridad_total/data.pickle")
df = df.transpose()
df.head()
```

```python

x = np.arange(0, len(df[0.500][0]))
d = df[0.500][0]
plt.plot(x,-d )
plt.show()
```

```python
xs, hs = sp.signal.find_peaks(-d, height=(-4.5, 10))
hs = -hs["peak_heights"]
plt.plot(x, d)
plt.plot(xs, hs, 'o', color="orange")
plt.show()
```

```python
bins = np.linspace(-10, 10, 50)
freq, bin_edges = np.histogram(d, bins=bins)
plt.stairs(freq, bin_edges, fill=True)
plt.yscale("log")
plt.show()
```
