```python
starting_wl = 370
ending_wl = 699
int_time = 0.05
wls = np.array(range(starting_wl, ending_wl))
spectrum = np.zeros(ending_wl - starting_wl, dtype=int)

for i, wl in enumerate(wls):
    counts = 0
    p = basepath/str(wl)
    for f in os.listdir(p):
        ch1 = pd.read_pickle(p/f).ch1
        counts += len(np.where(np.diff(ch1) < 0.5)[0])
    spectrum[i] = counts / int_time
```

```python
plt.close('all')
plt.plot(wls, spectrum)
plt.show()
```
